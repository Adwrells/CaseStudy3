```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('amit_raj_dev_final_dataset_G6_CS3_T3.csv')
```

```python
# --- Basic Overview ---
print('Shape of dataset:', df.shape)
print('\nColumns in dataset:\n', df.columns.tolist())
print('\nData Types:\n', df.dtypes)
print('\nMissing Values Summary:\n', df.isnull().sum())
```

```python
# --- Descriptive Statistics ---
print('\nSummary Statistics:\n', df.describe().T)
```

```python
numeric_df = df.copy()
```

```python
#mean median mode max min std deviation of all columns
for col in numeric_df.select_dtypes(include=[np.number]).columns:
    print(f'\nStatistics for {col}:')
    print(f'Mean: {numeric_df[col].mean()}')
    print(f'Median: {numeric_df[col].median()}')
    print(f'Mode: {numeric_df[col].mode()[0]}')
    print(f'Max: {numeric_df[col].max()}')
    print(f'Min: {numeric_df[col].min()}')
    print(f'Standard Deviation: {numeric_df[col].std()}')
```

```python
# --- Handle Missing Values (if any) ---
missing_ratio = df.isnull().mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=missing_ratio.head(15), y=missing_ratio.head(15).index)
plt.title('Top 15 Columns with Missing Values')
plt.xlabel('Proportion Missing')
plt.ylabel('Column Name')
plt.show()
```

```python
#Histogram for numerical columns
numeric_cols = numeric_df.select_dtypes(include=[np.number]).columns
numeric_df[numeric_cols].hist(bins=15, figsize=(15, 10))
plt.tight_layout(rect=[0, 0, 2, 2])
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
#Boxplot for outlier detection
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=numeric_df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#Correlation matrix heatmap of numerical features

# --- Correlation Matrix ---
num_df = df.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(14,10))
sns.heatmap(num_df.corr(), cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# --- Distribution of Key Metrics ---
cols_to_plot = ['Total Revenue', 'Net Income', 'Profit Margin', 
                'After Tax ROE', 'Total Assets', 'Total Liabilities']

plt.figure(figsize=(15, 8))

for i, col in enumerate(cols_to_plot, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
# --- Relationship Between Profitability Metrics ---
plt.figure(figsize=(8,6))
sns.scatterplot(x='Total Revenue', y='Profit Margin', data=df)
plt.title('Profit Margin vs Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Profit Margin (%)')
plt.show()
```

```python
# --- Trend Over Time (Example: Average ROE per Year) ---
roe_trend = df.groupby('For Year')['After Tax ROE'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.lineplot(x='For Year', y='After Tax ROE', data=roe_trend, marker='o')
plt.title('Average After-Tax ROE Over Years')
plt.xlabel('Year')
plt.ylabel('Average After-Tax ROE (%)')
plt.grid(True)
plt.show()
```

```python
# --- Company-Level Profit Comparison ---
top_profit = df.groupby('Ticker Symbol')['Net Income'].mean().nlargest(10).reset_index()
plt.figure(figsize=(10,5))
sns.barplot(x='Ticker Symbol', y='Net Income', data=top_profit, palette='Blues_d')
plt.title('Top 10 Companies by Average Net Income')
plt.xticks(rotation=45)
plt.show()
```

```python
# --- Debt vs Equity Analysis ---
plt.figure(figsize=(8,6))
sns.scatterplot(x='Total Liabilities', y='Total Equity', data=df, alpha=0.7)
plt.title('Liabilities vs Equity Across Companies')
plt.xlabel('Total Liabilities')
plt.ylabel('Total Equity')
plt.show()
```

```python
# --- Profit Margin Distribution by Year ---
plt.figure(figsize=(10,6))
sns.boxplot(x='For Year', y='Profit Margin', data=df)
plt.title('Profit Margin Distribution Over Years')
plt.xlabel('Year')
plt.ylabel('Profit Margin (%)')
plt.show()
```

```python
# Line Chart for Key Time Based Trends
roe_trend = df.groupby('For Year')['After Tax ROE'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.lineplot(x='For Year', y='After Tax ROE', data=roe_trend, marker='o')
plt.title('Average After-Tax ROE Over Years')
plt.xlabel('Year')
plt.ylabel('Average After-Tax ROE (%)')
plt.grid(True)
plt.show()
```

```python
# Plot first 10 columns for clarity
for i, col in enumerate(df.select_dtypes(include=[np.number]).columns[:46], 1):
    plt.figure(figsize=(8,4))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Time series analysis of key financial metrics over years with legends
metrics = ['Total Revenue', 'Net Income', 'Profit Margin', 'After Tax ROE']
for metric in metrics:
    trend = df.groupby('For Year')[metric].mean().reset_index()
    plt.figure(figsize=(8,5))
    sns.lineplot(x='For Year', y=metric, data=trend, marker='o', label=metric)
    plt.title(f'Average {metric} Over Years')
    plt.xlabel('Year')
    plt.ylabel(f'Average {metric}')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# --- Summary Insights ---
print('\n--- Insights Summary ---')
print('* The dataset spans multiple years and companies with financial indicators like Revenue, Profit Margin, ROE, and Liabilities.')
print('* Profit Margin and ROE vary widely, with some extreme outliers (highly profitable or loss-making firms).')
print('* Total Revenue and Profit Margin have a weak positive correlation â€” large revenue does not always imply higher margin.')
print('* Average After-Tax ROE shows year-to-year variability, suggesting changing market or industry conditions.')
print('* A few companies dominate Net Income, showing skewness in profitability distribution.')
```
